# Ising模型相变与磁滞回线模拟

## 项目简介
本项目通过Metropolis蒙特卡洛算法研究二维Ising模型的相变现象与磁滞回线。

## 1. 初始化与导入

```python
import numpy as np
import matplotlib.pyplot as plt
from ising_model import IsingModel
from monte_carlo import MetropolisMonteCarlo
from hysteresis import simulate_hysteresis_loop
from visualization import plot_spin_configuration, plot_phase_transition, plot_hysteresis_loop

print("所有模块导入成功！")
```

## 2. 自旋构型展示

```python
# 不同温度下的自旋构型
size = 32
temperatures = [1.0, 2.27, 3.5]

fig, axes = plt.subplots(1, 3, figsize=(18, 6))

for i, T in enumerate(temperatures):
    model = IsingModel(size)
    mc = MetropolisMonteCarlo(model, T)
    mc.thermalize(steps=300)
    
    ax = plot_spin_configuration(model.spins, ax=axes[i], title=f'自旋构型 (T={T})')

plt.tight_layout()
plt.show()
```

## 3. 相变曲线模拟

```python
# 温度扫描
size = 32
T_list = np.linspace(1.5, 3.5, 15)
m_avg_list = []
c_list = []
chi_list = []

for T in T_list:
    print(f'正在模拟 T = {T:.2f}')
    model = IsingModel(size)
    mc = MetropolisMonteCarlo(model, T)
    energy_list, magnetization_list = mc.run_simulation(thermal_steps=200, measure_steps=100)
    
    m_avg_list.append(np.mean(np.abs(magnetization_list)))
    c_list.append(model.calculate_specific_heat(energy_list, T))
    chi_list.append(model.calculate_susceptibility(magnetization_list, T))

print("模拟完成！")
```

```python
# 绘制相变曲线
fig = plot_phase_transition(T_list, m_avg_list, c_list, chi_list)
plt.show()
```

## 4. 磁滞回线模拟

```python
# 磁滞回线模拟（T < Tc）
size = 32
T = 1.5

print(f'正在模拟磁滞回线 (T={T})...')
h_list, m_list = simulate_hysteresis_loop(size, T, h_min=-2.0, h_max=2.0, 
                                          num_h=30, thermal_steps=150, measure_steps=80)

fig = plot_hysteresis_loop(h_list, m_list, num_h=30, T=T)
plt.show()
```

## 5. 结论

- 相变温度 $T_c \approx 2.27$，与Onsager解析解一致
- 当 $T < T_c$ 时，系统呈现铁磁性，存在自发磁化
- 当 $T > T_c$ 时，系统处于顺磁相
- 磁滞回线是铁磁材料的重要特性，反映了系统对历史的记忆
